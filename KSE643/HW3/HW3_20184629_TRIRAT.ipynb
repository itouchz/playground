```python
import string
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

stop_words = set(stopwords.words('english'))

# Load file
filename = 'metamorphosis_clean.txt'
file = open(filename, 'rt', encoding='utf-8-sig')
text = file.read()
file.close()

# Split into words
tokens = word_tokenize(text)

# Convert to lower case
tokens = [w.lower() for w in tokens]

# Remove punctuation from each word
table = str.maketrans('', '', string.punctuation)
stripped = [w.translate(table) for w in tokens]

# Remove remaining tokens that are not alphabetic
words = [word for word in stripped if word.isalpha()]

# Filter out stop words
words = [w for w in words if not w in stop_words]

# Stemming of words
porter = PorterStemmer()
stemmed = [porter.stem(word) for word in words]
```

## Question 1
For the final set of terms (stems), compute their TF (term frequency) scores.

```python
from collections import Counter
from nltk.text import TextCollection

def compute_tf(word_dict, n_terms):
    tf = {}
    for word, count in word_dict.items():
        tf[word] = count/n_terms
    return tf

word_count = Counter(stemmed)
tf_scores = compute_tf(word_count, len(stemmed))
```

## Question 2
Rearrange the terms by their TF scores and print the top 50 words and their respective TF scores in pairs.

```python
import pandas as pd

tf_scores_df = pd.DataFrame.from_dict(tf_scores, orient='index', columns=['TF'])
tf_top_50 = tf_scores_df.sort_values(by='TF', ascending=False).head(50)
tf_top_50
```

```python
new_stop_words = stop_words

new_stop_words.add('mr')
new_stop_words.add('nt')
new_stop_words.add('one')
new_stop_words.add('could')
new_stop_words.add('would')

# Load file
filename = 'metamorphosis_clean.txt'
file = open(filename, 'rt', encoding='utf-8-sig')
text = file.read()
file.close()

# Split into words
tokens = word_tokenize(text)

# Convert to lower case
tokens = [w.lower() for w in tokens]

# Remove punctuation from each word
table = str.maketrans('', '', string.punctuation)
stripped = [w.translate(table) for w in tokens]

# Remove remaining tokens that are not alphabetic
words = [word for word in stripped if word.isalpha()]

# Filter out stop words
words = [w for w in words if not w in new_stop_words]

# Stemming of words
porter = PorterStemmer()
new_stemmed = [porter.stem(word) for word in words]
```

```python
word_count = Counter(new_stemmed)
new_tf_scores = compute_tf(word_count, len(new_stemmed))
```
